```python
import pandas as pd
import numpy as np
df = pd.read_excel('Ch3.ClevelandData.xlsx')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# BOXPLOT
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplot : ")
plt.show()
```

```python
# DATA CLEANING:
# missing values check krne ke liye
print(df.isnull().sum())

# missing values htane ke liye
df.dropna(inplace=True)

# non-numeric values htane ke liye
df.replace('?', np.nan, inplace=True)

# HEATMAP CODE:
plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlations")
plt.show()
```

```python
sns.pairplot(df)
plt.title("Scatter Matrix of the Dataset")
plt.show()
```
